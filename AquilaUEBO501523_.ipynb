```python

my_integer=10
my_float=3.5
my_string="Hello"
my_boolean=True

print(my_integer,"is of type:",type(my_integer))
print(my_float,"is of type:",type(my_float))
print(my_string,"is of type:",type(my_string))
print(my_boolean ,"is of type:",type(my_boolean))

float_val=19.99
int_val=int(float_val)
print(int_val,"is of type:",type(int_val))

int_val=50
str_val=str(int_val)
print(str_val,"is of type:",type(str_val))

int_val=50
str_val=str(int_val)
print(str_val,"is of type:",type(str_val))

first_name=input("Enter Aquila:")
last_name=input("Enter Vuore:")

print("Hello,",first_name+"!")
print(f"Hello,(first_name)(last_name)!")


age=20
print("You are "+str(age)+"year old.")

favourite_word=input("freedom")
num_repetitions=int(input("How many times do you want to repeat it?"))
ExceptValueError: print("invalid.Please enter a number.")
num_repetition=0
if num_repetitions>0:print((favourite_word + "")*num_repetitions)

a. print("Hello" + abc) - NameError (abc is not defined)
b. int("abc") - ValueError
c. age + 10 -TypeError




def to_lowercase(s):return s.lower()
print(to_lowercase("Hello"))
print(to_lowercase("here"))
print(to_lowercase("LOVELY"))

def swap_case(s):return s,swapcase0
print(swap_case("HeLLo WoRLD"))

def remove_uppercase(s):
    result = ""
    for char in s:
        if not char.isupper():
            result += char
    return result

print(remove_uppercase("HelloWorlD"))

Task

def count_upper_lower(s):
    upper_count = 0
    lower_count = 0
    for char in s:
        if char.isupper():
            upper_count += 1
        elif char.islower():
            lower_count += 1
    return upper_count, lower_count

upper, lower = count_upper_lower("EngiNEEER")
print("Uppercase:", upper, "Lowercase:", lower)

Task

import re

def remove_non_letters(s):
    return re.sub(r'[^a-zA-Z]', '', s)

print(remove_non_letters("Data-Driven@2025!"))


Task

import math

def triangle_area(a, b, c):
    s = (a + b + c) / 2
    area = math.sqrt(s * (s - a) * (s - b) * (s - c))
    return area

print(triangle_area(3, 4, 5))

Task

def format_names(names):
    for name in names:
        print(name.ljust(20), name.rjust(20), name.center(20))

names = ["Alice", "Bob", "Charlie", "David"]
format_names(names)

Task

def clean_string(s):
    s = s.strip()
    s = s.replace(" ", "")
    return s

print(clean_string("  Hello, World!  "))






Task
Variable_Creation

Create variables of these types: Integer, Float, String, and Boolean.
integer_var = 10
float_var = 19.99
string_var = "Hello, World!"
boolean_var = True


 Type__Conversion

* Assignment : Convert a float (19.99) to an integer.
     int_from_float = int(float_var)

* Assignment b: Convert an integer (50) to a string.
     str_from_int = str(50)

* Assignment c: Convert a string ("50") to a float.
     float_from_str = float("50")


 User Input and Greeting

 a: Write a program that takes the users first name and last name as input.
    first_name = input("Enter your first name: ")
    last_name = input("Enter your last name: ")

* Assignment b: Print a greeting message.
     print(f"Hello, {first_name} {last_name}!")
Task 4: Error Correction
* Original Code:
        age = 20
    print("You are " + age + " years old.")


* Correction: Convert age to a string.
        print("You are " + str(age) + " years old.")


# ---- Base Class ----
class PetroleumFormula:
    def calculate(self):
        raise NotImplementedError("Subclasses must implement this method.")

# ---- 1. Hydrostatic Pressure ----
class HydrostaticPressure(PetroleumFormula):
    def _init_(self, mw, tvd):
        self.mw = mw
        self.tvd = tvd

    def calculate(self):
        try:
            return 0.052 * self.mw * self.tvd
        except Exception as e:
            print(f"Error calculating Hydrostatic Pressure: {e}")

# ---- 2. Porosity ----
class Porosity(PetroleumFormula):
    def _init_(self, pore_volume, bulk_volume):
        self.pore_volume = pore_volume
        self.bulk_volume = bulk_volume

    def calculate(self):
        try:
            return self.pore_volume / self.bulk_volume
        except ZeroDivisionError:
            print("Error: Bulk volume cannot be zero.")

# ---- 3. Permeability (Darcy’s Law) ----
class Permeability(PetroleumFormula):
    def _init_(self, k, A, mu, L, p1, p2):
        self.k = k
        self.A = A
        self.mu = mu
        self.L = L
        self.p1 = p1
        self.p2 = p2

    def calculate(self):
        try:
            return (self.k * self.A * (self.p1 - self.p2)) / (self.mu * self.L)
        except ZeroDivisionError:
            print("Error: Viscosity and length cannot be zero.")

# ---- 4. Formation Volume Factor ----
class FormationVolumeFactor(PetroleumFormula):
    def _init_(self, vol_res, vol_stb):
        self.vol_res = vol_res
        self.vol_stb = vol_stb

    def calculate(self):
        try:
            return self.vol_res / self.vol_stb
        except ZeroDivisionError:
            print("Error: STB volume cannot be zero.")

# ---- 5. API Gravity ----
class APIGravity(PetroleumFormula):
    def _init_(self, sg):
        self.sg = sg

    def calculate(self):
        try:
            return (141.5 / self.sg) - 131.5
        except ZeroDivisionError:
            print("Error: Specific gravity cannot be zero.")

# ---- 6. Gas Compressibility Factor ----
class GasCompressibility(PetroleumFormula):
    def _init_(self, P, V, n, R, T):
        self.P = P
        self.V = V
        self.n = n
        self.R = R
        self.T = T

    def calculate(self):
        try:
            return (self.P * self.V) / (self.n * self.R * self.T)
        except ZeroDivisionError:
            print("Error: n, R, and T must be non-zero.")

if _name_ == "_main_":

    # Create different formula objects
    formulas = [
        HydrostaticPressure(12, 10000),        # Mud weight = 12 ppg, TVD = 10000 ft
        Porosity(200, 1000),                   # Vp = 200, Vb = 1000
        Permeability(100, 50, 1.2, 10, 3000, 1000),  # k, A, mu, L, p1, p2
        FormationVolumeFactor(1.5, 1.0),       # Reservoir oil / Stock tank oil
        APIGravity(0.85),                      # SG = 0.85
        GasCompressibility(2000, 10, 2, 10.73, 520)  # P, V, n, R, T
    ]

    # Polymorphism in action: same method name "calculate" works for all
    print("=== Petroleum Engineering Calculations ===")
for f in formulas:
        print(f"{f._class.name_}: {f.calculate()}")


class PetroleumFormula:
    def calculate(self):
        raise NotImplementedError("Subclasses must implement this method.")

 1. Hydrostatic Pressure
class HydrostaticPressure(PetroleumFormula):
    def _init_(self, mw, tvd):
        self.mw = mw
        self.tvd = tvd

    def calculate(self):
        try:
            return 0.052 * self.mw * self.tvd
        except Exception as e:
            print(f"Error calculating hydrostatic pressure: {e}")

 2. Porosity
class Porosity(PetroleumFormula):
    def _init_(self, pore_volume, bulk_volume):
        self.pore_volume = pore_volume
        self.bulk_volume = bulk_volume

    def calculate(self):
        try:
            return self.pore_volume / self.bulk_volume
        except ZeroDivisionError:
            print("Error: Bulk volume cannot be zero.")

 3. Permeability (Darcy’s Law)
class Permeability(PetroleumFormula):
    def _init_(self, k, A, mu, L, p1, p2):
        self.k = k; self.A = A; self.mu = mu; self.L = L
        self.p1 = p1; self.p2 = p2

    def calculate(self):
        try:
            return (self.k * self.A * (self.p1 - self.p2)) / (self.mu * self.L)
        except ZeroDivisionError:
            print("Error: Viscosity and length cannot be zero.")

 4. Formation Volume Factor
class FormationVolumeFactor(PetroleumFormula):
    def _init_(self, vol_res, vol_stb):
        self.vol_res = vol_res
        self.vol_stb = vol_stb

    def calculate(self):
        try:
            return self.vol_res / self.vol_stb
        except ZeroDivisionError:
            print("Error: STB volume cannot be zero.")

5. API Gravity
class APIGravity(PetroleumFormula):
    def _init_(self, sg):
        self.sg = sg

    def calculate(self):
        try:
            return (141.5 / self.sg) - 131.5
        except ZeroDivisionError:
            print("Error: Specific gravity cannot be zero.")

 6. Real Gas Compressibility Factor
class GasCompressibility(PetroleumFormula):
    def _init_(self, P, V, n, R, T):
        self.P = P; self.V = V; self.n = n; self.R = R; self.T = T

    def calculate(self):
        try:
            return (self.P * self.V) / (self.n * self.R * self.T)
        except ZeroDivisionError:
            print("Error: n, R, and T must be non-zero.")


 Demonstrating Polymorphism
formulas = [
    HydrostaticPressure(12, 10000),
    Porosity(200, 1000),
    Permeability(100, 50, 1.2, 10, 3000, 1000),
    FormationVolumeFactor(1.5, 1.0),
    APIGravity(0.85),
    GasCompressibility(2000, 10, 2, 10.73, 520)]

for f in formulas:
    print(f"{f._class.name_}: {f.calculate()}")
```
